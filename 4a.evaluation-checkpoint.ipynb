## Validation data

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )

print (f'{DATADIR}/geopotential_500/')
valid_years = list(range( 2021, 2022+1))
print (valid_years)
z500_valid_files = [ file for year in valid_years for file in glob.glob (fr'{DATADIR}/geopotential/*{year}*') ] 
t850_valid_files = [ file for year in valid_years for file in glob.glob (fr'{DATADIR}/temperature/*{year}*')  ] 
z500_valid = load_test_data(z500_valid_files, 'z', slice('2021', '2022')).load()  
t850_valid = load_test_data(t850_valid_files, 't', slice('2021', '2022')).load() 

# for precipitation we are taking 6 hourly accumulations
tp_valid_files  = [ file for year in valid_years for file in glob.glob (fr'{DATADIR}/total_precipitation/*{year}*') ] 
t2m_valid_files = [ file for year in valid_years for file in glob.glob (fr'{DATADIR}/2m_temperature/*{year}*')  ] 
tp_valid = load_test_data(tp_valid_files,   'tp', slice('2021', '2022')).rolling(time=6).sum()
tp_valid.name = 'tp'
t2m_valid = load_test_data(t2m_valid_files, 't2m', slice('2021', '2022'))

valid_data = xr.merge([z500_valid, t850_valid, tp_valid, t2m_valid])
```

## Load data

First let's load all the prediction files from the previous notebooks.

```python
# to speed things up, let's only evaluate every 6th hour
valid_data = valid_data.isel(time=slice(0, None, 6))
```

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )
save_prefix = 'NCI_tutorial' 
print (f'{PREDDIR}/{save_prefix}_persistence_{res}.nc')
print (f'{PREDDIR}/{save_prefix}_climatology_{res}.nc')
persistence = xr.open_dataset(f'{PREDDIR}/{save_prefix}_persistence_{res}.nc', chunks={'time': 10}).load()
climatology = xr.open_dataset(f'{PREDDIR}/{save_prefix}_climatology_{res}.nc', chunks={'time': 10}).load()
weekly_climatology = xr.open_dataset(f'{PREDDIR}/{save_prefix}_weekly_climatology_{res}.nc', chunks={'time': 10}).load()
```

### Load linear regression checkpoints

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )
variables = ['z', 't', 'tp', 't2m']
print('Linear regression - 3d - Merging data ...')
[ print (f'{PREDDIR}/{save_prefix}_lr_3d_{v}_{v}.nc') for v in variables]
lr_3d = xr.merge([xr.open_dataset(f'{PREDDIR}/{save_prefix}_lr_3d_{v}_{v}.nc', chunks={'time': 10}) for v in variables]).load()
print('Linear regression - 5d - Merging data ...')
[ print (f'{PREDDIR}/{save_prefix}_lr_5d_{v}_{v}.nc') for v in variables]
lr_5d = xr.merge([xr.open_dataset(f'{PREDDIR}/{save_prefix}_lr_5d_{v}_{v}.nc', chunks={'time': 10}) for v in variables]).load()
print('Linear regression - 6h, iterative loading ...')
print(f'{PREDDIR}/{save_prefix}_lr_6h_iter.nc')
lr_6h_iter = xr.open_dataset(f'{PREDDIR}/{save_prefix}_lr_6h_iter.nc', chunks={'time': 10}).load()
```

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )
print ('save_prefix:', save_prefix)
cnn_3d = xr.open_dataset(f'{PREDDIR}/{save_prefix}_cnn_3d.nc', chunks={'time': 10})
cnn_5d = xr.open_dataset(f'{PREDDIR}/{save_prefix}_cnn_5d.nc', chunks={'time': 10})
cnn_6h_iter = xr.open_dataset(f'{PREDDIR}/{save_prefix}_fccnn_6h_iter.nc', chunks={'time': 10})
```

```python
variables = ['2m_temperature', 'total_precipitation']
```

## Compute RMSE

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )

func = compute_weighted_rmse
rmse = OrderedDict({
    'Persistence': evaluate_iterative_forecast(persistence, valid_data, func).load(),
    'Climatology': func(climatology, valid_data).load(),
    'Weekly clim.': func(weekly_climatology, valid_data).load(),
    'LR (iterative)': evaluate_iterative_forecast(lr_6h_iter, valid_data, func).load(),
    'CNN (iterative)': evaluate_iterative_forecast(cnn_6h_iter, valid_data, func).load(),
    'LR (direct)': xr.concat(
        [
            func(lr_3d, valid_data),
            func(lr_5d, valid_data)
        ],
        dim=pd.Index([72, 120], name='lead_time')
    ).load(),
    'CNN (direct)': xr.concat(
        [
            compute_weighted_rmse(cnn_3d, valid_data),
            compute_weighted_rmse(cnn_5d, valid_data)
        ],
        dim=pd.Index([72, 120], name='lead_time')
    ).load(),

})
```

## Compute ACC

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )

func = compute_weighted_acc
acc = OrderedDict({
    'Persistence': evaluate_iterative_forecast(persistence, valid_data, func).load(),
    'Climatology': func(climatology, valid_data).load(),
    'Weekly clim.': func(weekly_climatology, valid_data).load(),
    'LR (iterative)': evaluate_iterative_forecast(lr_6h_iter, valid_data, func).load(),
    'CNN (iterative)': evaluate_iterative_forecast(cnn_6h_iter, valid_data, func).load(),
    'LR (direct)': xr.concat(
        [
            func(lr_3d, valid_data),
            func(lr_5d, valid_data)
        ],
        dim=pd.Index([72, 120], name='lead_time')
    ).load(),
    'CNN (direct)': xr.concat(
        [
            func(cnn_3d, valid_data),
            func(cnn_5d, valid_data)
        ],
        dim=pd.Index([72, 120], name='lead_time')
    ).load(),

})
```

## Compute MAE

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )

func = compute_weighted_mae
mae = OrderedDict({
    'Persistence': evaluate_iterative_forecast(persistence, valid_data, func).load(),
    'Climatology': func(climatology, valid_data).load(),
    'Weekly clim.': func(weekly_climatology, valid_data).load(),
    'LR (iterative)': evaluate_iterative_forecast(lr_6h_iter, valid_data, func).load(),
    'CNN (iterative)': evaluate_iterative_forecast(cnn_6h_iter, valid_data, func).load(),
    'LR (direct)': xr.concat(
        [
            func(lr_3d, valid_data),
            func(lr_5d, valid_data)
        ],
        dim=pd.Index([72, 120], name='lead_time')
    ).load(),
    'CNN (direct)': xr.concat(
        [
            func(cnn_3d, valid_data),
            func(cnn_5d, valid_data)
        ],
        dim=pd.Index([72, 120], name='lead_time')
    ).load(),

})
```

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )
print ('save_prefix:', save_prefix)
print (f'{PREDDIR}/{save_prefix}_rmse.pkl')
print (f'{PREDDIR}/{save_prefix}_acc.pkl')
print (f'{PREDDIR}/{save_prefix}_mae.pkl')    
to_pickle(rmse, f'{PREDDIR}/{save_prefix}_rmse.pkl')
to_pickle(acc,  f'{PREDDIR}/{save_prefix}_acc.pkl')
to_pickle(mae,  f'{PREDDIR}/{save_prefix}_mae.pkl')
```

# The same procedure for higher resolutions

To see if the results are consistant for higher resolutions.

```python
res = '2.8125'
DATADIR = f'/g/data/wb00/NCI-Weatherbench/{res}deg' 
print(DATADIR)
PREDDIR = f"/scratch/vp91/{os.environ['USER']}/NCI-Weatherbench/pred_dir"
print(PREDDIR)
```

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )

valid_years = list(range( 2021, 2022+1))
print (valid_years)
z500_valid_files = [ file for year in valid_years for file in glob.glob (fr'{DATADIR}/geopotential/*{year}*') ] 
t850_valid_files = [ file for year in valid_years for file in glob.glob (fr'{DATADIR}/temperature/*{year}*')  ] 
z500_valid = load_test_data(z500_valid_files, 'z', slice('2021', '2022')).load()  
t850_valid = load_test_data(t850_valid_files, 't', slice('2021', '2022')).load() 
valid_data = xr.merge([z500_valid, t850_valid])
```

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )

save_prefix = 'NCI_tutorial' 
print ('save_prefix:', save_prefix)
persistence = xr.open_dataset(f'{PREDDIR}/{save_prefix}_persistence_{res}.nc', chunks={'time': 10}).load()#)
climatology = xr.open_dataset(f'{PREDDIR}/{save_prefix}_climatology_{res}.nc', chunks={'time': 10}).load()#)
weekly_climatology = xr.open_dataset(f'{PREDDIR}/{save_prefix}_weekly_climatology_{res}.nc', chunks={'time': 10}).load()#)
```

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )

rmse_persistence = evaluate_iterative_forecast(persistence, valid_data, compute_weighted_rmse).load()
rmse_climatology = compute_weighted_rmse(climatology, valid_data).load()
rmse_weekly_climatology = compute_weighted_rmse(weekly_climatology, valid_data).load()
```

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )

rmse = xr.merge(
    [r.rename({'z': f'z_{n}', 't': f't_{n}'}) for r, n in zip(
        [rmse_persistence, rmse_climatology, rmse_weekly_climatology #, 
        ],
        ['persistence', 'climatology', 'weekly_climatology' #,
        ]
    )],
)
```

```python
print ('save_prefix:', save_prefix)
rmse.to_netcdf(f'{PREDDIR}/{save_prefix}_rmse_{res}.nc')
```

```python
%%time
print( f'[{datetime.now().replace(microsecond=0)}]' )

for var in rmse:
    if 'lead_time' in rmse[var].coords:
        print(var, rmse[var].sel(lead_time=[3*24, 5*24]).values)
```
